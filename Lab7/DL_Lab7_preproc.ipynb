```python
import pandas as pd
df = pd.read_csv("/root/.cache/kagglehub/datasets/lakshmi25npathi/imdb-dataset-of-50k-movie-reviews/versions/1/IMDB Dataset.csv")
df
```

```python
import re
def preproc(x):
    x = re.sub(r"<br\s*/?\s*>","", x)
    x = re.sub(r"[^a-zA-Z0-9\s]", "", x)
    x = x.lower()
    return x
df["review"] = df["review"].apply(preproc)
df
```

```python
corpus = " ".join(df["review"])
len(corpus)
```

```python
from collections import Counter

def BPE(text):
    c = list(text)
    corpus = []
    for i in c:
        if i != " ":
            corpus.append(i)
    vocabulary = list(set(corpus))
    print(f"Vocabulary size: {len(vocabulary)}")
    j = 0
    while j < 200:
        print(j)
        pairs = [(corpus[i], corpus[i + 1]) for i in range(len(corpus) - 1)]
        pair_dict = Counter(pairs)

        if not pair_dict:
            break

        most_common1, most_common2 = pair_dict.most_common(1)[0][0]
        new_token = most_common1 + most_common2
        merged_corpus = []
        i = 0
        while i < len(corpus):
            if i < len(corpus) - 1 and corpus[i] == most_common1 and corpus[i + 1] == most_common2:
                merged_corpus.append(new_token)
                i += 2
            else:
                merged_corpus.append(corpus[i])
                i += 1

        corpus = merged_corpus
        if new_token not in vocabulary:
            vocabulary.append(new_token)

        if not any(token in corpus for token in vocabulary):
            break
        j += 1
    return vocabulary

vocab = BPE(corpus)
len(vocab)
```

```python
import json
with open("vocab.json", "w") as f:
    json.dump(vocab, f)
```

```python
def encode_word(word, vocab):
    encoded_word = [char for char in word if char != ' ']

    merge_dict = {token: vocab[token] for token in vocab if len(token) > 1}

    i = 0
    while i < len(encoded_word) - 1:
        pair = encoded_word[i] + encoded_word[i + 1]
        if pair in merge_dict:
            encoded_word[i] = pair  # Merge
            del encoded_word[i + 1]  # Remove second char
            i = max(0, i - 1)  # Step back to check for new merge
        else:
            i += 1
    encoded_ids = [vocab[token] for token in encoded_word]

    return encoded_ids
```

```python
vocab = {i : idx for idx, i in enumerate(vocab)}
```

```python
df["review1"] = df["review"].apply(lambda x: encode_word(x, vocab))
df
```
