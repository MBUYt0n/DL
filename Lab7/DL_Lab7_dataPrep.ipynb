```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shusrith/imbd-dataset-tokenized")

print("Path to dataset files:", path)
```

```python
import pandas as pd
import ast

df = pd.read_csv("/root/.cache/kagglehub/datasets/shusrith/imbd-dataset-tokenized/versions/1/data.csv")
df["review1"] = df["review1"].apply(ast.literal_eval)
```

```python
df["sentiment"].replace({"positive": 1, "negative": 0}, inplace=True)
df.drop(columns=["review"], inplace=True)
df
```

```python
max_len = 200
l = []
for i in range(len(df)):
    if len(df.iloc[i, 1]) > max_len:
        x = df.iloc[i, 1][200:]
        for j in range(len(x) // 200):
            if j == len(x) // 2:
                l.append([df.iloc[i, 0], x[200 * (j + 1) :]])
            else:
                l.append([df.iloc[i, 0], x[200 * (j + 1) : 200 * (j + 2)]])
```

```python
len(l)
```

```python
df1 = pd.DataFrame(l, columns=["sentiment", "review1"])
df1
```

```python
df1 = pd.concat([df, df1], ignore_index=True)
```

```python
df1.to_csv("data.csv")
```
